```python
import matplotlib.pyplot as plt
date=["25/2","26/2","27/2"]
temp=[8.5,10.5,6.8]
plt.plot(date,temp)
plt.show()
```

```python
import matplotlib.pyplot as plt
date=["25/2","26/2","27/2"]
temp=[8.5,10.5,6.8]
plt.plot(date,temp,marker='p',linestyle="dashdot",color="red",linewidth="2")
plt.show()
```

```python
date=['25/2','4/2','30/7','15/7','24/9']
temp=[8.3,13.0,20.0,34.34,45.32]
plt.plot(date,temp)
plt.xlabel('date',color='blue',size=10)
plt.ylabel('tempreture',color='blue',size=10)
plt.title("date vs temp",loc='center',color='green',size=15)
plt.grid(True)
plt.yticks([10,20,30,40,50])
#plt.xticks(visible=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]
max_temp=[78,97,87,85,68,48,75]
min_temp=[54,56,25,35,15,42,51]
ave_temp=[54,45,65,52,54,65,58]
plt.plot(days,max_temp,label='max')
plt.plot(days,min_temp,label="min")
plt.plot(days,ave_temp,label='ave')
plt.xlabel('weekdays',color='green',size=15)
plt.ylabel('tempreture',color='green',size=15)
plt.title('weather week',color='red',size=20)
#plt.annotate(xy=[3,56],text="its3,56")
plt.legend(loc='best',shadow=True,fontsize='small')
plt.grid(True)
```
