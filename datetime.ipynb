```python
#convert timedate to its string
#using str()
import datetime as dt
now=dt.datetime.now()
s=str(now)
print(s)
```

```python
#using strftime()
year=now.strftime("%Y")
print("year:",year)
```

```python
month=now.strftime("%m")
print("month:",month)
week=now.strftime("%U")
print("week:",week)
day=now.strftime("%d")
print("day:",day)
mon=now.strftime("%h")
print("mon:",mon)
hour=now.strftime("%H")
print("hour:",hour)
minute=now.strftime("%M")
print("minute:",minute)
second=now.strftime("%S")
print("second:",second)
```

```python
#time transformation
#time data
import datetime as dt
now=dt.datetime.now()
t=now+dt.timedelta(days=50,seconds=2341,microseconds=2324,milliseconds=3567890,minutes=5,hours=6,weeks=9)
print("current time:",now.strftime("%d,%Y,%H,%M,%S,%m"))
print("time transformation:",t.strftime("%d,%Y,%H,%M,%S,%m"))
print("time diffrence:",t-now)
```
